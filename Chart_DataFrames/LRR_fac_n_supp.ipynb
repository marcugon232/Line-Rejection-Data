```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
supp = pd.read_csv("LRR_supplier.csv")
supp
```

```python
plt.bar(supp["SUPPLIER"],supp["LRR (ppm)"], width=0.8, bottom=None, align='center');
plt.title("Supplier vs. LRR (ppm)");
plt.xlabel("Supplier");
plt.ylabel("LRR (ppm)");
plt.savefig("SupplierVLRR(bar).png")
```

```python
fac = pd.read_csv("LRR_factory.csv")
fac
```

```python
plt.bar(fac["FACTORY"],fac["LRR (ppm)"], width=0.8, bottom=None, align='center');
plt.title("Factory vs. LRR (ppm)");
plt.xlabel("Factory");
plt.ylabel("LRR (ppm)");
plt.savefig("SupplierVLRR(bar).png")
```
