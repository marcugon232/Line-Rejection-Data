```python
import pandas as pd
import matplotlib.pyplot


```

```python
LRRxFact = pd.read_csv("LRR_factory.csv")
LRRxFact
```

```python
fac_mon = LRRxFact.groupby(["FACTORY","MONTH"]).sum()
fac_mon = fac_mon.reset_index()
fac_mon
```

```python
plt.scatter(fac_mon["MONTH"],fac_mon["FACTORY"],c = "Black", alpha = .75, s = fac_mon["LRR (ppm)"]);
plt.title("Month Vs. Factory\n Size of Circle Correlates to LRR", fontsize = 14)
plt.xlabel("Months", fontsize = 12)
plt.ylabel("Cities", fontsize = 12)
plt.savefig("MonthxFacLRR.png")
```

```python
supp_mon = pd.read_csv("LRR_supplier.csv")
plt.scatter(supp_mon["MONTH"],supp_mon["SUPPLIER"],c = "Black", alpha = .75, s = supp_mon["LRR (ppm)"]);
plt.title("Month Vs. Supplier\n Size of Circle Correlates to LRR", fontsize = 14)
plt.xlabel("Months", fontsize = 12)
plt.ylabel("Cities", fontsize = 12)
plt.savefig("MonthxSuppLRR.png")
```
