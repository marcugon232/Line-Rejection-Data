```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
LRR_factory= pd.read_csv("LRR_factory(pie).csv")
LRR_factory.head()
```

```python
LRR_sum = LRR_factory["LRR (ppm)"].sum()

LRR_factory["Percent"]=LRR_factory["LRR (ppm)"]/LRR_sum * 100
LRR_factory
```

```python
factory = ["Brasilia", "Chengdu", "Mumbai", "Taoyuan"]
lrr_per = ["32.933017", "8.265866", "51.793736", "7.007381"]
colors =  ["yellow", "green", "lightblue", "orange","red", "purple"]
explode = (0.1, 0, 0, 0)
```

```python
plt.pie(lrr_per, explode=explode, labels=factory, colors=colors,autopct="%1.1f%%", shadow=True, startangle=140);
plt.savefig("LRRfactory(pie).png")
```
